```python
# ðŸ“¦ Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ðŸ“š (Assume these functions already exist from your model file)
from HandiCapperAccuracyModel import fetch_player_game_logs, compute_player_distribution, STAT_COLUMNS

# ðŸ§ª Load Tyler Herro's data
player_name = "Tyler Herro"

# Fetch raw game logs
df_herro = fetch_player_game_logs(player_name)
display(df_herro.head())

# Build distribution model
herro_model = compute_player_distribution(player_name)

if herro_model:
    mean_vector = herro_model["mean_vector"]
    covariance_matrix = herro_model["covariance_matrix"]

    # ðŸ“ˆ Display mean stats
    print(f"ðŸ“Š Average Stats for {player_name}")
    mean_df = pd.DataFrame({
        "Stat": STAT_COLUMNS,
        "Average Value": mean_vector
    })
    display(mean_df)

    # ðŸ§® Display covariance matrix
    print(f"ðŸ§® Covariance Matrix for {player_name}")
    cov_df = pd.DataFrame(covariance_matrix, index=STAT_COLUMNS, columns=STAT_COLUMNS)
    display(cov_df)

    # ðŸŽ¨ Plot Heatmap of Covariance
    plt.figure(figsize=(10, 8))
    sns.heatmap(cov_df, annot=True, cmap="coolwarm", fmt=".2f", square=True)
    plt.title(f"Covariance Heatmap - {player_name}", fontsize=16)
    plt.show()

    # ðŸŽ¨ Plot Heatmap of Correlation
    corr_df = df_herro.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_df, annot=True, cmap="viridis", fmt=".2f", square=True)
    plt.title(f"Correlation Heatmap - {player_name}", fontsize=16)
    plt.show()

else:
    print(f"âŒ Could not build model for {player_name} (probably not enough games)")
```
