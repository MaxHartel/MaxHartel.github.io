```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import combinations
from HandiCapperAccuracyModel import fetch_player_game_logs

# Constants
STAT_COLUMNS = ["points", "rebounds", "assists", "blocks", "steals", "turnovers", "minutes"]

def compute_team_distribution(player_names, league, connection):
    # Fetch all players' logs and concatenate
    dfs = []
    for name in player_names:
        df = fetch_player_game_logs(name, league, connection)
        if df is not None and not df.empty:
            dfs.append(df[STAT_COLUMNS])
    
    if not dfs:
        print("âŒ No valid data found for any players.")
        return

    df_team = pd.concat(dfs, ignore_index=True)

    # Compute covariance and correlation
    cov_matrix = df_team.cov()
    corr_matrix = df_team.corr()

    # Top 5 stat pairs by absolute covariance
    cov_pairs = [
        (i, j, cov_matrix.loc[i, j])
        for i, j in combinations(cov_matrix.columns, 2)
    ]
    top_cov = sorted(cov_pairs, key=lambda x: abs(x[2]), reverse=True)[:5]

    # Top 5 stat pairs by absolute correlation
    corr_pairs = [
        (i, j, corr_matrix.loc[i, j])
        for i, j in combinations(corr_matrix.columns, 2)
    ]
    top_corr = sorted(corr_pairs, key=lambda x: abs(x[2]), reverse=True)[:5]

    # ðŸ§¾ Print top results
    print("ðŸ“ˆ Top 5 Stat Pairs by Covariance:")
    for a, b, val in top_cov:
        print(f" - {a} & {b}: {val:.2f}")

    print("\nðŸ”— Top 5 Stat Pairs by Correlation:")
    for a, b, val in top_corr:
        print(f" - {a} & {b}: {val:.2f}")

    # ðŸŽ¨ Visualize Covariance
    sns.heatmap(cov_matrix, annot=True, cmap="coolwarm", fmt=".2f", square=True)
    plt.title("Team Covariance Heatmap")
    plt.show()

    # ðŸŽ¨ Visualize Correlation
    sns.heatmap(corr_matrix, annot=True, cmap="viridis", fmt=".2f", square=True)
    plt.title("Team Correlation Heatmap")
    plt.show()

team = ["Tyler Herro", "Davion Mitchell", "Bam Adebayo"]
league = "NBA"
conn = get_connection()

compute_team_distribution(team, league, conn)
```

```python
def rename_player_stats(df, player_name, stat_columns):
    """
    Renames stat columns to include the player's name for disambiguation
    when combining data from multiple players.

    Args:
        df (pd.DataFrame): DataFrame of a single player's stats.
        player_name (str): Name of the player (e.g., "Tyler Herro").
        stat_columns (list): List of stat columns to rename (e.g., ['points', 'assists']).

    Returns:
        pd.DataFrame: A copy of the DataFrame with renamed stat columns.
    """
    renamed_df = df.copy()
    rename_map = {
        col: f"{player_name}'s {col.replace('_', ' ').title()}" for col in stat_columns
    }
    return renamed_df.rename(columns=rename_map)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
from HandiCapperAccuracyModel import fetch_player_game_logs

# âœ… Define which stats to include
STAT_COLUMNS = [
    'points', 'rebounds', 'assists', 'steals', 'blocks', 'turnovers',
    'personal_fouls', 'threes_made', 'minutes', 'pa', 'pr', 'ra', 'pra', 'fantasy_points'
]

# âœ… Rename stat columns to "Player's Stat"
def rename_player_stats(df, player_name, stat_columns):
    renamed_df = df.copy()
    rename_map = {
        col: f"{player_name}'s {col.replace('_', ' ').title()}" for col in stat_columns
    }
    return renamed_df.rename(columns=rename_map)

# âœ… Merge multiple player DataFrames on game_date
def merge_players_stats(players, league, connection):
    merged_df = None
    for player in players:
        df = fetch_player_game_logs(player, league, connection)
        df = rename_player_stats(df, player, STAT_COLUMNS)
        if merged_df is None:
            merged_df = df
        else:
            merged_df = pd.merge(merged_df, df, on='game_date', how='inner')
    return merged_df

# âœ… Display top N absolute correlations or covariances between stat pairs
# def top_stat_pairs(matrix, top_n=5, metric='Covariance'):
#     pairs = []
#     for i, j in combinations(matrix.columns, 2):
#         val = matrix.loc[i, j]
#         pairs.append(((i, j), val))

#     # Sort by absolute value
#     sorted_pairs = sorted(pairs, key=lambda x: abs(x[1]), reverse=True)[:top_n]
#     print(f"\nðŸ” Top {top_n} Stat Pairs by {metric}:")
#     for (stat1, stat2), value in sorted_pairs:
#         print(f"  {stat1} & {stat2}: {value:.2f}")

# âœ… Display top N absolute correlations or covariances between stat pairs of 2 different players
def top_stat_pairs(matrix, top_n=15, metric='Covariance'):
    pairs = []
    for i, j in combinations(matrix.columns, 2):
        # Skip pairs from the same player
        player_i = i.split("'")[0].strip()
        player_j = j.split("'")[0].strip()
        if player_i == player_j:
            continue

        val = matrix.loc[i, j]
        pairs.append(((i, j), val))

    # Sort by absolute value
    sorted_pairs = sorted(pairs, key=lambda x: abs(x[1]), reverse=True)[:top_n]

    print(f"\nðŸ” Top {top_n} Stat Pairs by {metric} (Different Players Only):")
    for (stat1, stat2), value in sorted_pairs:
        print(f"  {stat1} & {stat2}: {value:.2f}")

# âœ… Master analysis function
def analyze_multiplayer_stats(players, league, connection):
    df = merge_players_stats(players, league, connection)

    # Drop non-stat columns
    stat_cols = [col for col in df.columns if any(stat in col.lower() for stat in STAT_COLUMNS)]
    df_stats = df[stat_cols].dropna()

    # Compute covariance and correlation
    cov_matrix = df_stats.cov()
    corr_matrix = df_stats.corr()

    # Display results
    print("\nðŸ§® Covariance Matrix")
    display(pd.DataFrame(cov_matrix))

    print("\nðŸ”— Correlation Matrix")
    display(pd.DataFrame(corr_matrix))

    # Show heatmaps
    plt.figure(figsize=(12, 10))
    sns.heatmap(cov_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
    plt.title("Covariance Matrix Across Player Stats", fontsize=14)
    plt.show()

    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="viridis", square=True)
    plt.title("Correlation Matrix Across Player Stats", fontsize=14)
    plt.show()

    # Top correlated and covariant stat pairs
    top_stat_pairs(cov_matrix, metric='Covariance')
    top_stat_pairs(corr_matrix, metric='Correlation')

from planet_scale_port import get_connection
conn = get_connection()
analyze_multiplayer_stats(["Tyler Herro", "Bam Adebayo", "Davion Mitchell"], league="NBA", connection=conn)
```
